```python
def reverse_string(string):
    return string[::-1]
```

```python
reverse_string("Python")
```

```python
reverse_string("hello")
```

```python
def reverse_string_manual(string):
    reversed_string = ""
    for char in string:
        reversed_string = char + reversed_string
    return reversed_string
print(reverse_string_manual("This is a test."))
```

```python
import string as str
def reverse_string_manual_no_spaces(string):
    reversed_string = ""
    for char in string:
        if char != " " and char not in str.punctuation:
            reversed_string = char + reversed_string
    return reversed_string
reverse_string_manual_no_spaces("This is a test!?!")
```

```python
import string as str
def reverse_string_manual_for_lists(string):
    reversed_string = ""
    for char in string:
        if char != " " and char not in str.punctuation:
            reversed_string = char + reversed_string
    return reversed_string
reverse_string_manual_for_lists(["eggs", "milk", "bread", "butter"])
```

```python
import string as str
def reverse_string_manual_list_reverse(user_input):
    string = ' '.join(user_input)
    reversed_string = ""
    for char in string:
        if char != " " and char not in str.punctuation:
            reversed_string = char + reversed_string
    return reversed_string
reverse_string_manual_list_reverse(["eggs", "milk", "bread", "butter"])
# reverse_string_manual_list_reverse("This is a test!?!?")
```
