```python
def find_max(numbers):
    if numbers is None or len(numbers) == 0:
        return ("No numbers were provided!")
    max_number = numbers[0]
    for number in numbers:
        if max_number > number:
            continue
        else:
            max_number = number
            max_index = numbers.index(number)
    return f"Max number is {max_number} and Index is [{max_index}]"

find_max([31, 12, 48, 15, 59, 99, 23, 64, 53, 37, 55])
#find_max([9, 3, 8, -4, -2, 5, 0])
```
