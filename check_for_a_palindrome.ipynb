```python
import string as str
def is_palindrome(string):
    string = string.lower().replace(" ", "")
    final_string = ""
    for char in string:
        if char not in str.punctuation:
            final_string = final_string + char
    if final_string == final_string[::-1]:
        return "Yes, it's a palindrome!"
    else:
        return "No, it's not a palindrome."

#is_palindrome("racecar")
is_palindrome("taco! cat!")
#is_palindrome("Hello")
#is_palindrome("A man a plan a canal Panama")
```

```python
import re
def is_palindrome(string):
    clean = re.sub(r'[^\w]', '', string).lower()
    if clean == clean[::-1]:
        return "Yes, it's a palindrome!"
    else:
        return "No, it's not a palindrome."

#is_palindrome("racecar")
is_palindrome("taco! cat!")
#is_palindrome("Hello")
#is_palindrome("A man a plan a canal Panama")
```
