```python
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

is_prime(19)
```

```python
def all_prime(number):
    primes = []
    for num in range(2, number + 1):
        if is_prime(num):
            primes.append(num)
    return primes

all_prime(19)
```
